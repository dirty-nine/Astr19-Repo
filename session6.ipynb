import necessary dependencies

```python
import numpy as np
```

returns the sin of the argument in radians as type numpy.float64

```python
def sin(x):
    return np.sin(x)
```

returns the cos of the argument in radians as type numpy.float64

```python
def cos(x):
    return np.cos(x)
```

set dataTable to the entire tabulated list of sin(x) and cos(x) vs. x where x is tabulated between 0 and 2pi with a thousand entries

```python
x = np.linspace(0, 2*np.pi, 1000)
sinx, cosx = sin(x), cos(x)
dataTable = np.transpose((sinx,cosx,x))
```

prints out the first 10 values of x, sin(x), and cos(x) in columns

```python
for i in range(0,10):
    print(dataTable[i])
```

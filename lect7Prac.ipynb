```python
import  numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# %matplotlib inline
```

```python
n = 20

x1 = np.linspace(-2*np.pi,2*np.pi, n)
y1 = np.sin(x1)

x2 = x1
y2 = np.log10(x2)
```

```python
# plt.plot(x,y)
#make points that will become lines on the graph
plt.plot(x1, y1, marker='o', color = 'g', label='l1')
plt.plot(x2, y2, color = "blue", marker='o', linestyle='--', label='l2')
plt.plot(x1, y1+1, 'ro', label = 'l3')

#create legend
plt.legend(loc=0)

#create axis
plt.xlabel('x')
plt.ylabel('y')

#save graph as file and show in jupyter notebook
plt.savefig('lec7Plot1.png', bbox_inches='tight', dpi=600)
plt.show()
```

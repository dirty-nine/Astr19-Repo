```python
import numpy as np
import matplotlib.pyplot as plt    
```

```python
def scatter_plot(x, y, z, zerr,
                 flag_save=True,
                 xlabel='x',
                 ylabel='y',
                 lcolor='red',
                 pcolor='gold',
                 fs=14,
                 fname='scatter_plot.png'):

    f, ax = plt.subplots(1,1,figsize=(4,4))

    ax.plot(x, y, color=lcolor, linewidth=1.5, label='model (line)')

    ax.errorbar(x, z, yerr=zerr,
                fmt='o',
                color=pcolor,
                markersize=3,
                label='data (points)')

    ax.set_xlabel(xlabel, fontsize=fs)
    ax.set_ylabel(ylabel, fontsize=fs)

    ax.legend(loc='upper left', frameon=False)

    if flag_save:
        plt.savefig(fname, bbox_inches='tight', dpi=400)
```

```python
def main():

    print('Making a plot!')

    x = np.linspace(0,1,10)
    y = x**2

    z = y + np.random.normal(scale=0.1, size=len(x))
    zerr = np.full_like(x, 0.1)

    scatter_plot(x, y, z, zerr)
```
